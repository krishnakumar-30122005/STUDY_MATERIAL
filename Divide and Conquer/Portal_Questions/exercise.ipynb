```python
n=int(input())
arr=list(map(int,input().split()))
def merge(arr,l,m,r):
    left=arr[l:m+1]
    right=arr[m+1:r+1]
    i=j=0
    k=l
    while i<len(left) and j<len(right):
        if left[i]<=right[j]:
            arr[k]=left[i]
            i+=1
        else:
            arr[k]=right[j]
            j+=1
        k+=1
    while i<len(left):
        arr[k]=left[i]
        i+=1
        k+=1
    while j<len(right):
        arr[k]=right[j]
        j+=1
        k+=1
    print(*arr)
def merge_sort(arr,l,r):
    if l<r:
        m=(l+r)//2
        merge_sort(arr,l,m)
        merge_sort(arr,m+1,r)
        merge(arr,l,m,r)
merge_sort(arr,0,n-1)
```

![Max_and_Min](Images/Max_and_Min.png)

```python
def minimum(arr):
    if len(arr)==1:
        return arr[0],arr[0]
    mid=len(arr)//2
    left=arr[:mid]
    right=arr[mid:]
    left_min,left_max=minimum(left)
    right_min,right_max=minimum(right)
    final=min(left_min,right_min)
    final1=max(left_max,right_max)
    return final,final1
n=int(input())
arr=list(map(int,input().split(",")))
min,max=minimum(arr)
print("Minimum:",min)
print("Maximum:",max)
```

![skyline_problem](Images/skyline_prblm.png)

```python
buildings = eval(input())
def divide(left,right):
    if left == right:
        l,r,h = buildings[left]
        return [[l,h],[r,0]]
    mid = (left+right)//2
    left_sky = divide(left,mid)
    right_sky = divide(mid+1,right)
    
    return merge(left_sky,right_sky)

def merge(left,right):
    i = j = 0
    h1 = h2 = 0
    result = []
    
    while i<len(left) and j<len(right):
        if left[i][0] < right[j][0]:
            x,h1 = left[i]
            i+=1
        elif left[i][0] > right[j][0]:
            x,h2 = right[j]
            j+=1
        else:
            x,h1 = left[i]
            h2 = right[j][1]
            i+=1
            j+=1
        max_h = max(h1,h2)
        if not result or result[-1][1]!=max_h:
            result.append([x,max_h])
    while i<len(left):
        if not result or result[-1][1] != left[i][1]:
            result.append([left[i][0],left[i][1]])
            i+=1
    while j<len(right):
        if not result or result[-1][1] != right[j][1]:
            result.append([right[j][0],right[j][1]])
            j+=1
    return result
result = divide(0,len(buildings)-1)
ans =[]
for points in result:
    ans.append(tuple(points))
print(ans)
    
```
