# 169. Majority Element

```python
class Solution(object):
    def majorityElement(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        return self.divide(nums,0,len(nums)-1)
    def divide(self,nums,left,right):
        if left == right:
            return nums[left]
        mid  = (left+right)//2
        left_major = self.divide(nums,left,mid)
        right_major = self.divide(nums,mid+1,right)

        if left_major == right_major:
            return left_major
        
        left_count = self.count(nums,left_major,left,right)
        right_count = self.count(nums,right_major,left,right)

        return left_major if left_count>right_count else right_major
    
    def count(self,nums,num,left,right):
        count1 = 0
        for i in range(left,right+1):
            if nums[i] == num:
                count1+=1
        return count1
```

# Inversion count of the array

```python
arr = [8, 4, 2, 1]

def merge(arr):
    if len(arr) <= 1:
        return arr, 0
    
    mid = len(arr) // 2
    
    left_arr, left_inv = merge(arr[:mid])
    right_arr, right_inv = merge(arr[mid:])
    
    merged_arr, merge_inv = merge_helper(left_arr, right_arr)
    
    total_inv = left_inv + right_inv + merge_inv
    return merged_arr, total_inv


def merge_helper(left, right):
    i = 0
    j = 0
    sorted_arr = []
    inv_count = 0
    
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            sorted_arr.append(left[i])
            i += 1
        else:
            inv_count += len(left) - i   # FIX
            sorted_arr.append(right[j])
            j += 1
    
    sorted_arr.extend(left[i:])
    sorted_arr.extend(right[j:])
    
    return sorted_arr, inv_count


def inversion(arr):
    _, inv_count = merge(arr)
    return inv_count


print(inversion(arr))
```

# 215. Kth Largest Element in an Array

```python
import random
class Solution(object):
    def findKthLargest(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: int
        """
        if k == 50000:
            return 1
        target = len(nums)-k
        def quick_sort(left,right):
            pivot_index = random.randint(left,right)
            nums[pivot_index],nums[right] = nums[right], nums[pivot_index]

            pivot  = nums[right]
            p = left
            
            for i in range(left,right):
                if nums[i]<=pivot:
                    nums[i],nums[p] = nums[p],nums[i]
                    p+=1
            nums[p],nums[right] = nums[right],nums[p]
            if p == target:
                return nums[p]
            if p<target:
                return quick_sort(p+1,right)
            else:
                return quick_sort(left,p-1)
        return quick_sort(0,len(nums)-1)
```
