# 4. Median of Two Sorted Arrays

```python
class Solution(object):
    def findMedianSortedArrays(self, nums1, nums2):
        """
        :type nums1: List[int]
        :type nums2: List[int]
        :rtype: float
        """
        i=0
        j=0
        total = len(nums1)+len(nums2)
        sorted_arr = []
        while i<len(nums1) and j<len(nums2):
            if nums1[i]<=nums2[j]:
                sorted_arr.append(nums1[i])
                i+=1
            else:
                sorted_arr.append(nums2[j])
                j+=1
        sorted_arr.extend(nums1[i:])
        sorted_arr.extend(nums2[j:])
        mid = total//2
        if total%2==1:
            return sorted_arr[total//2]
        else:
            return (sorted_arr[mid-1] + sorted_arr[mid]) / 2.0



```

# 23. Merge k Sorted Lists

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
import heapq
class Solution(object):
    def mergeKLists(self, lists):
        """
        :type lists: List[Optional[ListNode]]
        :rtype: Optional[ListNode]
        """
        
        heap = []
        dummy = ListNode(0)
        curr = dummy
        for i, node in enumerate(lists):
            if node:
                heapq.heappush(heap,(node.val,i,node))
        while heap:
            val,i,node = heapq.heappop(heap)
            curr.next = node
            if node.next:
                heapq.heappush(heap,(node.next.val,i,node.next))
            curr = curr.next
        return dummy.next
```

# 148. Sort List

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def sortList(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: Optional[ListNode]
        """
        def get_mid(head):
            prev = None
            slow = head
            fast = head

            while fast and fast.next:
                prev = slow
                slow = slow.next
                fast = fast.next.next
            if prev:
                prev.next = None
            return slow

        def merge(slow,fast):
            dummy = ListNode(0)
            curr = dummy

            while slow and fast:
                if slow.val<=fast.val:
                    curr.next = slow
                    slow = slow.next
                else:
                    curr.next = fast
                    fast = fast.next
                curr = curr.next
            curr.next = fast if fast else slow
            return dummy.next
        def merge_sort(head):
            if head == None or head.next == None:
                return head
            mid = get_mid(head)
            left_part = merge_sort(head)
            right_part = merge_sort(mid)

            return merge(left_part,right_part)
        
        return merge_sort(head)
```

# 315. Count of Smaller Numbers After Self

```python
class Solution(object):
    def countSmaller(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        n = len(nums)
        nums = list(enumerate(nums))
        result = [0]*n
        def inversion(nums):
            merged_arr,inv_count = merge_sort(nums)
            return merged_arr,inv_count
        def merge_sort(arr):
            if len(arr)<=1:
                return arr,0
            mid = len(arr)//2
            left_part,left_inv = merge_sort(arr[:mid])
            right_part,right_inv = merge_sort(arr[mid:])

            merge_arr,merge_inv = merge(left_part,right_part)

            return merge_arr, left_inv+right_inv+merge_inv

        def merge(left,right):
            i = 0
            j = 0
            inv_count = 0
            sorted_arr = []
            while i<len(left) and j<len(right):
                if left[i][1] <= right[j][1]:
                    result[left[i][0]]+=j
                    sorted_arr.append(left[i])
                    i+=1
                else:
                    inv_count+= len(left)-i
                    sorted_arr.append(right[j])
                    j+=1
                
            while i<len(left):
                result[left[i][0]]+=j
                sorted_arr.append(left[i])
                i+=1
            while j<len(right):
                sorted_arr.append(right[j])
                j+=1
                
            return sorted_arr , inv_count
        _,inversion_count = inversion(nums)

        return result
```

# 327. Count of Range Sum

```python
class Solution(object):
    def countRangeSum(self, nums, lower, upper):
        """
        :type nums: List[int]
        :type lower: int
        :type upper: int
        :rtype: int
        """
        prefix= [0]
        for i in nums:
            prefix.append(prefix[-1]+i)
    
        def merge_sort(arr):
            if len(arr)<=1:
                return arr,0
            mid = len(arr)//2
            left,left_count = merge_sort(arr[:mid])
            right,right_count = merge_sort(arr[mid:])

            count = left_count + right_count

            j = k = 0
            for val in left:
                while j<len(right) and right[j]-val<lower:
                    j+=1
                while k<len(right) and right[k]-val<=upper:
                    k+=1
                count += k-j
        
            sorted_arr = []
            i = j = 0
            while i< len(left) and j<len(right):
                if left[i] <= right[j]:
                    sorted_arr.append(left[i])
                    i+=1
                else:
                    sorted_arr.append(right[j])
                    j+=1
            sorted_arr.extend(left[i:])
            sorted_arr.extend(right[j:])

            return sorted_arr,count
        _,count = merge_sort(prefix)
        return count
            
        
        
```

# 493. Reverse Pairs

```python
class Solution(object):
    def reversePairs(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        def merge_sort(arr):
            if len(arr)<=1:
                return arr,0
            mid = len(arr)//2
            left,left_count = merge_sort(arr[:mid])
            right,right_count = merge_sort(arr[mid:])

            result = left_count+right_count
            j=0
            for i in range(len(left)):
                while j<len(right) and left[i]>2*right[j]:
                    j+=1
                result+=j
            
            sorted_arr = []
            i = j = 0

            while i<len(left) and j<len(right):
                if left[i]<=right[j]:
                    sorted_arr.append(left[i])
                    i+=1
                else:
                    sorted_arr.append(right[j])
                    j+=1
                
            sorted_arr.extend(left[i:])
            sorted_arr.extend(right[j:])

            return sorted_arr,result
        _,result = merge_sort(nums)
        return result
        


```
