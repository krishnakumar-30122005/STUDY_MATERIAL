# Divide and conquer Algorithm

Divide and Conquer algorithm is a problem-solving strategy that involves.

Divide : Break the given problem into smaller non-overlapping problems.

Conquer : Solve Smaller Problems

Combine : Use the Solutions of Smaller Problems to find the overall result.

Examples of Divide and Conquer are Merge Sort, Quick Sort, Binary Search and Closest Pair of Points

# Merge Sort

```python
class Solution(object):
    def merge_sort(self,arr):
        if len(arr)<=1:
            return arr
        mid = len(arr)//2
        left = self.merge_sort(arr[:mid])
        right = self.merge_sort(arr[mid:])

        return self.merge(left,right)
    def sortArray(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        return self.merge_sort(nums)

    def merge(self,left,right):
        sorted_arr=[]

        i=j=0
        while i< len(left) and j<len(right):
            if left[i]<right[j]:
                sorted_arr.append(left[i])
                i+=1
            else:
                sorted_arr.append(right[j])
                j+=1
        sorted_arr.extend(left[i:])
        sorted_arr.extend(right[j:])

        return sorted_arr
    
```

# LAMUTO PARTITION

```python
class Solution(object):
    def sortColors(self, nums):
        """
        :type nums: List[int]
        :rtype: None Do not return anything, modify nums in-place instead.
        """
        self.quick_sort(nums,0,len(nums)-1)
    def partition(self,arr,left,right):
        pivot = arr[right]
        i = left-1

        for j in range(left,right):
            if arr[j]< pivot:
                i+=1
                arr[i],arr[j] = arr[j],arr[i]
        arr[i+1],arr[right] = arr[right],arr[i+1]
        return i+1
    def quick_sort(self,arr,low,high):
        if low<high:
            pi = self.partition(arr,low,high)

            self.quick_sort(arr,low,pi-1)
            self.quick_sort(arr,pi+1,high)

        return arr
        
```

# NAIVE PARTITION


```python
def partition(arr):
    n = len(arr)
    i = 0
    pivot = arr[n-1]
    temp = [0]*n

    for j in range(0,n):
        if arr[j]<=pivot:
            temp[i] = arr[j]
            i+=1
        
    for j in range(n):
        if arr[j]>pivot:
            temp[i] = arr[j]
            i+=1

    for j in range(n):
        arr[j] = temp[j]
    return arr
print(partition([1,6,4,3,2]))
    
```

# HOARE's Partition

```python
def partition(arr, low, high):
    pivot = arr[low]
    i = low - 1
    j = high + 1

    while True:
        while True:
            i += 1
            if arr[i] >= pivot:
                break

        while True:
            j -= 1
            if arr[j] <= pivot:
                break

        if i >= j:
            return j

        arr[i], arr[j] = arr[j], arr[i]
```

# Binary Search 

```python
def search(arr,target):
    n = len(arr)
    i=0
    j=n-1

    while i<=j:
        mid = (i+j)//2
        if arr[mid] == target:
            return mid
        elif arr[mid]<target:
            i = mid+1
        else:
            j = mid-1
    return -1
print(search([1,2,3,4,5],4))
```

# Minimum Distance between Two Points

```python
import math

def distance(p1, p2):
    return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)


def brute_force(points):
    min_dist = float('inf')
    n = len(points)

    for i in range(n):
        for j in range(i + 1, n):
            min_dist = min(min_dist, distance(points[i], points[j]))

    return min_dist


# Main recursive divide function
def closest_pair_recursive(points):
    n = len(points)

    # Base case
    if n <= 3:
        return brute_force(points)

    # Divide
    mid = n // 2
    mid_x = points[mid][0]

    left_points = points[:mid]
    right_points = points[mid:]

    # Conquer
    dl = closest_pair_recursive(left_points)
    dr = closest_pair_recursive(right_points)
    d = min(dl, dr)

    # Build strip
    strip = []
    for p in points:
        if abs(p[0] - mid_x) < d:
            strip.append(p)

    # Sort strip by y-coordinate
    strip.sort(key=lambda x: x[1])

    # Compare each point with next 6 points
    for i in range(len(strip)):
        for j in range(i + 1, min(i + 7, len(strip))):
            d = min(d, distance(strip[i], strip[j]))

    return d


# Wrapper function
def closest_pair(points):
    # Sort points by x-coordinate
    points = sorted(points, key=lambda x: x[0])
    return closest_pair_recursive(points)


# ------------------ TEST ------------------
points = [(2, 3), (12, 30), (40, 50), (5, 1), (12, 10), (3, 4)]
print("Minimum Distance:", closest_pair(points))
```

# Inversion of a Array

```python
def print_inversions(arr):

    def solve(problem):
        # Base case
        if len(problem) <= 1:
            return problem, []

        # Divide
        mid = len(problem) // 2
        left_part = problem[:mid]
        right_part = problem[mid:]

        # Conquer
        left_arr, left_inv = solve(left_part)
        right_arr, right_inv = solve(right_part)

        # Combine
        merged_arr, cross_inv = combine(left_arr, right_arr)

        return merged_arr, left_inv + right_inv + cross_inv

    def combine(left, right):
        i = j = 0
        merged = []
        inversions = []

        while i < len(left) and j < len(right):
            if left[i] <= right[j]:
                merged.append(left[i])
                i += 1
            else:
                merged.append(right[j])
                # PRINT ALL inversion pairs
                for k in range(i, len(left)):
                    inversions.append((left[k], right[j]))
                j += 1

        merged.extend(left[i:])
        merged.extend(right[j:])

        return merged, inversions

    _, inv_pairs = solve(arr)
    return inv_pairs
arr = [5, 2, 6, 1]
print(print_inversions(arr))

  
```
