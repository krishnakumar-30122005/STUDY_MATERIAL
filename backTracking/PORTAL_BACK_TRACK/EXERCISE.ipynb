![Knight_tour](Images/Knight_tour.png)

```python
num1 = int(input())
sx = int(input())
sy = int(input())
tx = int(input())
ty = int(input())

ans = [float('inf')]

def min_knight(start,target,n):
    visited =[[False]*(n+1) for _ in range(n+1)]
    sx,sy = start
    directions = [(1,2),(-1,2),(1,-2),(-1,-2),(2,1),(-2,1),(2,-1),(-2,-1)]
    def dfs(x,y,steps):
        if steps>=ans[0]:
            return 
        if (x,y) == target:
            ans[0] =steps
            return
        
        for x1,y1 in directions:
            n1,n2 = x+x1,y+y1
            if 1<=n1<=n and 1<=n2<=n and not visited[n1][n2]:
                visited[n1][n2]= True
                dfs(n1,n2,steps+1)
                visited[n1][n2] = False
    visited[sx][sy] = True
    dfs(sx,sy,0)
    
    return ans[0] if ans[0]!=float('inf') else  -1
    
min_knight((sx,sy),(tx,ty),num1)

if ans[0] == float('inf'):
    print("No solution")
else:
    print("Minimum steps =", ans[0])
```

![can_reach](Images/can_reach.png)

```python
num1 = 8
board = []

sx = sy = 0

for i in range(8):
    board.append(input().split(","))

for i in range(8):
    for j in range(8):
        if board[i][j] == "S":
            sx, sy = i, j
            break

def can_reach(start, n):
    visited = [[False]*n for _ in range(n)]
    directions = [(1,2),(-1,2),(1,-2),(-1,-2),
                  (2,1),(-2,1),(2,-1),(-2,-1)]

    def dfs(x, y):
        if board[x][y] == "X":
            return True

        for dx, dy in directions:
            nx, ny = x + dx, y + dy
            if 0 <= nx < n and 0 <= ny < n:
                if not visited[nx][ny] and board[nx][ny] != "1":
                    visited[nx][ny] = True 
                    if dfs(nx, ny):
                        return True
        return False

    visited[start[0]][start[1]] = True
    return dfs(start[0], start[1])

if can_reach((sx, sy), num1):
    print("YES")
else:
    print("NO")
```

![Hamiltonian_circuit](Images/Hamiltonian_circuit.png)

```python
num1 = int(input())
graph = []

for _ in range(num1):
    graph.append(list(map(int,input().split())))

path = [-1]*num1
visited = [False]*num1

path[0] = 0
visited[0] = True

ans = [False]
def solve(pos):
    if pos == num1:
        if graph[path[pos-1]][path[0]]==1:
            ans[0] = True
            return 
    for v in range(1,num1):
        if graph[path[pos-1]][v] == 1 and not visited[v]:
            visited[v] = True
            path[pos] = v
            solve(pos+1)
            if ans[0]:
                return 
            visited[v]  = False
            path[pos] =-1
solve(1)
if ans[0]:
    print("YES")
    print(*path,"0")
else:
    print("NO")
```

```python
n = int(input())

board = [[0]*n for _ in range(n)]
col = set()
sr = set()
sc = set()
ans= [0]

def solve(r):
    if r == n:
        ans[0]+=1
        return
    for c in range(n):
        if c in col or r+c in sr or r-c in sc:
            continue
        col.add(c)
        sr.add(r+c)
        sc.add(r-c)
        solve(r+1)
        col.remove(c)
        sr.remove(r+c)
        sc.remove(r-c)
    return 
solve(0)
print(ans[0])
    
```
