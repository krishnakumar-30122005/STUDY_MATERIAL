```python
n = int(input())

sx,sy = 0,0
tx,ty = n-1,n-1

grid = []

for _ in range(n):
  grid.append(input().split())


solution = [[0]*n for _ in range(n)]

if grid[sx][sy] == "1" or grid[tx][ty] == "1":
  print("No solution exists")
  exit()
  
path =[]
ans = [float('inf')]
def solve(x,y,step):
  if x==tx and y == ty:
    if step<ans[0]:
      solution[tx][ty] = 1
      ans[0] = step
      path.clear()
      path.append([row[:] for row in solution])
      
  if step>=ans[0]:
    return 

  if x<0 or y<0 or x>=n or y>=n or grid[x][y] == "1":
    return 
  
  solution[x][y] = 1
  solve(x+1,y,step+1)
  solve(x,y+1,step+1)
  solution[x][y] = 0
  return

solve(sx,sy,0)

for row in path[0]:
  print(*row)
```

```python
num1 = int(input())
sx = int(input())
sy = int(input())
tx = int(input())
ty = int(input())

ans = [float('inf')]

def min_knight(start,target,n):
    visited =[[False]*(n+1) for _ in range(n+1)]
    sx,sy = start
    directions = [(1,2),(-1,2),(1,-2),(-1,-2),(2,1),(-2,1),(2,-1),(-2,-1)]
    def dfs(x,y,steps):
        if steps>=ans[0]:
            return 
        if (x,y) == target:
            ans[0] =steps
            return
        
        for x1,y1 in directions:
            n1,n2 = x+x1,y+y1
            if 1<=n1<=n and 1<=n2<=n and not visited[n1][n2]:
                visited[n1][n2]= True
                dfs(n1,n2,steps+1)
                visited[n1][n2] = False
    visited[sx][sy] = True
    dfs(sx,sy,0)
    
    return ans[0] if ans[0]!=float('inf') else  -1
    
min_knight((sx,sy),(tx,ty),num1)

if ans[0] == float('inf'):
    print("No solution")
else:
    print("Minimum steps =", ans[0])
```

![can_reach](Images/can_reach.png)

```python
num1 = 8
board = []

sx = sy = 0

for i in range(8):
    board.append(input().split(","))

for i in range(8):
    for j in range(8):
        if board[i][j] == "S":
            sx, sy = i, j
            break

def can_reach(start, n):
    visited = [[False]*n for _ in range(n)]
    directions = [(1,2),(-1,2),(1,-2),(-1,-2),
                  (2,1),(-2,1),(2,-1),(-2,-1)]

    def dfs(x, y):
        if board[x][y] == "X":
            return True

        for dx, dy in directions:
            nx, ny = x + dx, y + dy
            if 0 <= nx < n and 0 <= ny < n:
                if not visited[nx][ny] and board[nx][ny] != "1":
                    visited[nx][ny] = True 
                    if dfs(nx, ny):
                        return True
        return False

    visited[start[0]][start[1]] = True
    return dfs(start[0], start[1])

if can_reach((sx, sy), num1):
    print("YES")
else:
    print("NO")
```
