# Candy- Leetcode 135

```python
class Solution(object):
    def candy(self, ratings):
        """
        :type ratings: List[int]
        :rtype: int
        """
        n = len(ratings)
        arr1  = [1]*n
        arr2  = [1]*n

        for i in range(1,n):
            if ratings[i-1]<ratings[i]:
                arr1[i] = arr1[i-1]+1
        for i in range(n-2,-1,-1):
            if ratings[i+1]<ratings[i]:
                arr2[i] = arr2[i+1]+1 
        candy = 0
        for i in range(n):
            candy+=(max(arr1[i],arr2[i]))

        return candy
```

# 241. Different Ways to Add Parentheses

```python
class Solution(object):
    def diffWaysToCompute(self, expression):
        """
        :type expression: str
        :rtype: List[int]
        """
        operations = {
            "+" : lambda x,y: x+y,
            "-" : lambda x,y: x-y,
            "*" : lambda x,y: x*y,
        }

        def backtrack(left,right):
            res = []
            for i in range(left,right+1):
                op = expression[i]
                if op in operations:
                    num1=backtrack(left,i-1)
                    num2=backtrack(i+1,right)

                    for n1 in num1:
                        for n2 in num2:
                            res.append(operations[op](n1,n2))
            if res == []:
                res.append(int(expression[left:right+1]))
            return res
        return backtrack(0,len(expression)-1)


        
```

# 2027. Minimum Moves to Convert String

```python
class Solution(object):
    def minimumMoves(self, s):
        """
        :type s: str
        :rtype: int
        """
        i = 0
        n = len(s)
        count = 0
        while i<n:
            if s[i] == "X":
                count+=1
                i+=3
            else:
                i+=1
        return count
```

# 169. Majority Element

```python
class Solution(object):
    def majorityElement(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        return self.divide(nums,0,len(nums)-1)
    def divide(self,nums,left,right):
        if left == right:
            return nums[left]
        mid  = (left+right)//2
        left_major = self.divide(nums,left,mid)
        right_major = self.divide(nums,mid+1,right)

        if left_major == right_major:
            return left_major
        
        left_count = self.count(nums,left_major,left,right)
        right_count = self.count(nums,right_major,left,right)

        return left_major if left_count>right_count else right_major
    
    def count(self,nums,num,left,right):
        count1 = 0
        for i in range(left,right+1):
            if nums[i] == num:
                count1+=1
        return count1
```

# 3. Longest Substring Without Repeating Characters

```python
class Solution(object):
    def lengthOfLongestSubstring(self, s):
        """
        :type s: str
        :rtype: int
        """
        ans = 0
        l=0
        seen = set()

        for r in range(len(s)):
            while s[r] in seen:
                seen.remove(s[l])
                l+=1
            seen.add(s[r])
            ans = max(ans,r-l+1)
        return ans
```

# 300. Longest Increasing Subsequence

```python
class Solution(object):
    def lengthOfLIS(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """

        n = len(nums)
        dp= [1]*n

        for i in range(n):
            for j in range(i):
                if nums[i]>nums[j]:
                    dp[i] = max(dp[i],dp[j]+1)
        return max(dp)
```

# House Robber - LC198 

```python
class Solution(object):
    def rob(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """ 
        n = len(nums)
        dp= [0]*(n)
        if n ==1:
            return nums[0]
        dp[0] = nums[0]
        dp[1] = max(nums[0],nums[1])

        for i in range(2,n):
            dp[i] = max(dp[i-1],nums[i]+dp[i-2])

        return dp[-1]
        
```

# 2266. Count Number of Texts

```python
class Solution(object):
    def countTexts(self, pressedKeys):
        """
        :type pressedKeys: str
        :rtype: int
        """

        n = len(pressedKeys)
        dp = [0]*(n+1)
        MOD = 10**9 + 7
        dp[0] = 1
        for i in range(1,n+1):
            ch = pressedKeys[i-1]
            if ch == "7" or ch == "9":
                limit = 4
            else:
                limit = 3
            for j in range(1,limit+1):
                if i-j < 0 :
                    break
                if ch != pressedKeys[i-j]:
                    break
                dp[i] = dp[i]+dp[i-j]
            
            dp[i] = dp[i]%MOD
        print(dp)
        return dp[n]
```

# 139. Word Break

```python
class Solution(object):
    def wordBreak(self, s, wordDict):
        """
        :type s: str
        :type wordDict: List[str]
        :rtype: bool
        """
        n = len(s)
        dp = [False]*(len(s)+1)
        dp[0] = True
        for i in range(n+1):
            for j in range(i+1):
                if dp[j] and "".join(s[j:i]) in wordDict:
                    dp[i] = True
                    break
        return dp[-1]
        
```

# 841. Keys and Rooms

```python
from collections import deque
rooms = [[1,3],[3,0,1],[2],[0]]
visited = [False]*(len(rooms))
visited[0] = True
set1 = set()
set1.add(0)
queue = deque()
queue.append(rooms[0])

while queue:
    current_room = queue.popleft()
    for keys in current_room:
        if keys in set1:
            continue 
        set1.add(keys)
        visited[keys] = True
        queue.append(rooms[keys])
print(len(set1) == len(rooms))
```
