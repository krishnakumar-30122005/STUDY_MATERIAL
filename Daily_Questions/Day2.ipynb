# Candy- Leetcode 135

```python
class Solution(object):
    def candy(self, ratings):
        """
        :type ratings: List[int]
        :rtype: int
        """
        n = len(ratings)
        arr1  = [1]*n
        arr2  = [1]*n

        for i in range(1,n):
            if ratings[i-1]<ratings[i]:
                arr1[i] = arr1[i-1]+1
        for i in range(n-2,-1,-1):
            if ratings[i+1]<ratings[i]:
                arr2[i] = arr2[i+1]+1 
        candy = 0
        for i in range(n):
            candy+=(max(arr1[i],arr2[i]))

        return candy
```

# 241. Different Ways to Add Parentheses

```python
class Solution(object):
    def diffWaysToCompute(self, expression):
        """
        :type expression: str
        :rtype: List[int]
        """
        operations = {
            "+" : lambda x,y: x+y,
            "-" : lambda x,y: x-y,
            "*" : lambda x,y: x*y,
        }

        def backtrack(left,right):
            res = []
            for i in range(left,right+1):
                op = expression[i]
                if op in operations:
                    num1=backtrack(left,i-1)
                    num2=backtrack(i+1,right)

                    for n1 in num1:
                        for n2 in num2:
                            res.append(operations[op](n1,n2))
            if res == []:
                res.append(int(expression[left:right+1]))
            return res
        return backtrack(0,len(expression)-1)


        
```

# 2027. Minimum Moves to Convert String

```python
class Solution(object):
    def minimumMoves(self, s):
        """
        :type s: str
        :rtype: int
        """
        i = 0
        n = len(s)
        count = 0
        while i<n:
            if s[i] == "X":
                count+=1
                i+=3
            else:
                i+=1
        return count
```

# 169. Majority Element

```python
class Solution(object):
    def majorityElement(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        return self.divide(nums,0,len(nums)-1)
    def divide(self,nums,left,right):
        if left == right:
            return nums[left]
        mid  = (left+right)//2
        left_major = self.divide(nums,left,mid)
        right_major = self.divide(nums,mid+1,right)

        if left_major == right_major:
            return left_major
        
        left_count = self.count(nums,left_major,left,right)
        right_count = self.count(nums,right_major,left,right)

        return left_major if left_count>right_count else right_major
    
    def count(self,nums,num,left,right):
        count1 = 0
        for i in range(left,right+1):
            if nums[i] == num:
                count1+=1
        return count1
```

# Construct Binary Tree from Preorder and Inorder Traversal 105 

```python
class Solution(object):
    def buildTree(self, preorder, inorder):
        """
        :type preorder: List[int]
        :type inorder: List[int]
        :rtype: Optional[TreeNode]
        """
        if not preorder or not inorder:
            return None
        root = TreeNode(preorder[0])
        mid = inorder.index(preorder[0])
        root.left = self.buildTree(preorder[1:mid+1],inorder[:mid])
        root.right = self.buildTree(preorder[mid+1:],inorder[mid+1:])
        return root

```

# 3. Longest Substring Without Repeating Characters

```python
class Solution(object):
    def lengthOfLongestSubstring(self, s):
        """
        :type s: str
        :rtype: int
        """
        ans = 0
        l=0
        seen = set()

        for r in range(len(s)):
            while s[r] in seen:
                seen.remove(s[l])
                l+=1
            seen.add(s[r])
            ans = max(ans,r-l+1)
        return ans
```
