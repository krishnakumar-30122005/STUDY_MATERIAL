# Predict the Winner (486)

```python
class Solution(object):
    def predictTheWinner(self, nums):
        """
        :type nums: List[int]
        :rtype: bool
        """

        n = len(nums)
        dp = {}

        def solve(l,r):
            if l == r:
                return nums[l]
            if (l,r) in dp:
                return dp[(l,r)]
            
            pick_left = nums[l] - solve(l+1,r)
            pick_right = nums[r] - solve(l,r-1)


            dp[(l,r)] = max(pick_left,pick_right)
            return dp[(l,r)]
        return solve(0,n-1)>=0
        
```
