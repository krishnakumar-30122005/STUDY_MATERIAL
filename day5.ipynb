# LEETCODE 131
Palindrome Partitioning

```python
class Solution(object):
    def partition(self, s):
        """
        :type s: str
        :rtype: List[List[str]]
        """
        def palindrome(str1):
            return str1==str1[::-1]
        path =[]
        res =[]
        n = len(s)
        def solve(start):
            if start == n:
                path.append(res[:])
                return
            for i in range(start,n):
                if palindrome(s[start:i+1]):
                    res.append(s[start:i+1])
                    solve(i+1)
                    res.pop()
            return 
        solve(0)
        return path
                
```

# LEETCODE : 93 Restore IP Addresses

```python
s = "25525511135"
n = len(s)
path = []
res=[]

def isvalid(s):
    if len(s)==1:
        return True
    if s[0]=="0":
        return False
    if 0<=int(s)<=255:
        return True
def solve(start):
    if start == n:
        if len(res)==4:
            path.append(".".join(res))
        return 
    if len(path)>4:
        return
    for i in range(start,min(start+3,n)):
        if isvalid(s[start:i+1]):
            res.append(s[start:i+1])
            solve(i+1)
            res.pop()
    return 
solve(0)
print(path)
        
```

# 784. Letter Case Permutation

```python
s = "a1b2"
res =[]
path =[]
n = len(s)
def solve(idx):
    if idx == n:
        path.append(''.join(res))
        return 
    if s[idx].isdigit():
        res.append(s[idx])
        solve(idx+1)
        res.pop()
    else:
        res.append(s[idx].lower())
        solve(idx+1)
        res.pop()
        res.append(s[idx].upper())
        solve(idx+1)
        res.pop()
    return 
solve(0)
print(path)
```
