# LEETCODE 131
Palindrome Partitioning

```python
class Solution(object):
    def partition(self, s):
        """
        :type s: str
        :rtype: List[List[str]]
        """
        def palindrome(str1):
            return str1==str1[::-1]
        path =[]
        res =[]
        n = len(s)
        def solve(start):
            if start == n:
                path.append(res[:])
                return
            for i in range(start,n):
                if palindrome(s[start:i+1]):
                    res.append(s[start:i+1])
                    solve(i+1)
                    res.pop()
            return 
        solve(0)
        return path
                
```
