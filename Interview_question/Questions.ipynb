# 34. Find First and Last Position of Element in Sorted Array

```python
class Solution(object):
    def searchRange(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        def first(nums):
            i = 0
            j = len(nums)-1

            first = -1

            while i<=j:
                mid = (i+j)//2
                if nums[mid] == target:
                    first = mid
                    j = mid-1
                elif nums[mid]<target:
                    i = mid+1
                else:
                    j = mid-1
            return first
        def second(nums):
            i = 0
            j = len(nums)-1

            second = -1

            while i<=j:
                mid = (i+j)//2
                if nums[mid] == target:
                    second = mid
                    i = mid+1
                elif nums[mid]<target:
                    i = mid+1
                else:
                    j = mid-1
            return second
        return [first(nums),second(nums)]
```

# 1456. Maximum Number of Vowels in a Substring of Given Length

```python
class Solution(object):
    def maxVowels(self, s, k):
        """
        :type s: str
        :type k: int
        :rtype: int
        """
        vowels = "aeiou"
        count = 0
        for i in range(k):
            if s[i] in vowels:
                count+=1
        max_count = count
        n = len(s)
        for i in range(k,n):
            if s[i] in vowels:
                count+=1
            if s[i-k] in vowels:
                count-=1
            max_count = max(max_count,count)
        return max_count
```
