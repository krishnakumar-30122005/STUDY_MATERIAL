```python
n = int(input())
elements = list(map(int, input().split()))

hash_table = [-1] * n

def h1(x):
    return x % n

def h2(x):
    return 7 - (x % 7)

for x in elements:
    i = 0
    while True:
        index = (h1(x) + i * h2(x)) % n
        if hash_table[index] == -1:
            hash_table[index] = x
            break
        i += 1

for i in range(n):
    print(f"index {i} value = {hash_table[i]}")
```

```python
n = int(input())
elements = list(map(int, input().split()))

hash_table = [-1] * n

def hash_function(x, i):
    return (x % n + i * i) % n

for x in elements:
    i = 0
    while i < n:
        index = hash_function(x, i)
        if hash_table[index] == -1:
            hash_table[index] = x
            break
        i += 1
for i in range(n):
    print(f"Element at position {i}: {hash_table[i]}")
```

```python
n = int(input())
elements = list(map(int, input().split()))
hash_table = [-1] * n

for x in elements:
    index = x % n
    while hash_table[index] != -1:
        index = (index + 1) % n
    hash_table[index] = x

for i in range(n):
    print(f"index {i} value = {hash_table[i]}")
```

```python
n = int(input())
elements = list(map(int, input().split()))

hash_table = [[] for _ in range(n)]

for x in elements:
    index = x % n
    hash_table[index].append(x)

for i in range(n):
    print(f"at index {i}")
    if len(hash_table[i]) == 0:
        print("No Hash Entry")
    else:
        for val in hash_table[i]:
            print(f"{val}->", end="")
        print()
```
