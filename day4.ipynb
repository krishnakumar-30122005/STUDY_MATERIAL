# COMBINATION SUM

```python
class Solution(object):
    def combinationSum(self, candidates, target):
        """
        :type candidates: List[int]
        :type target: int
        :rtype: List[List[int]]
        """
        res =[]
        path = []
        n = len(candidates)
        total = 0
        def solve(start,total):
            if total == target:
                path.append(res[:])
            if total>target:
                return
            for i in range(start,n):
                res.append(candidates[i])
                solve(i,total+candidates[i])
                res.pop()
            return 
        solve(0,0)
        return path
```

```python
class Solution(object):
    def subsetsWithDup(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        res = []
        path =[]
        nums.sort()
        n = len(nums)
        def solve(start):
            path.append(res[:])
            for i in range(start,n):
                if i>start and nums[i]==nums[i-1]:
                    continue
                res.append(nums[i])
                solve(i+1)
                res.pop()
            return 
        solve(0)
        return path
        
```

# KNIGHT PROBLEM

```python
n = 8
sx,sy = 0,0
tx,ty = 7,7
ans=[float('inf')]
visited = [[False]*n for _ in range(n)]
def solve(x,y,step):
    if x==tx and y==ty:
        if ans[0]>step:
            ans[0] = step
        return 
    if x<0 or y<0 or x>=n or y>=n or visited[x][y]==True:
        return
    if step>=ans[0]:
        return
    visited[x][y] = True
    solve(x+2,y+1,step+1)
    solve(x-2,y-1,step+1)
    solve(x+1,y+2,step+1)
    solve(x-1,y-2,step+1)
    solve(x+2,y-1,step+1)
    solve(x-2,y+1,step+1)
    solve(x+1,y-2,step+1)
    solve(x-1,y+2,step+1)
    visited[x][y] = False
    return 
solve(0,0,0)
print(ans[0])

```

# N QUEEN

```python
col =set()
s1 = set()
s2 = set()
n = 4
grid=[["."]*n for _ in range(n)]
res =[]
def solve(r):
    if r == len(grid):
        res.append([''.join(x) for x in grid])
    for c in range(n):
        if c in col or r+c in s1 or r-c in s2:
            continue
        grid[r][c] ="Q"
        col.add(c)
        s1.add(r+c)
        s2.add(r-c)
        solve(r+1)
        grid[r][c]="."
        col.remove(c)
        s1.remove(r+c)
        s2.remove(r-c)
    return
solve(0)
for row in res:
    print(*row)
    print("_________________________________________________")
```

# SURROUNDED REGION

```python
class Solution(object):
    def solve(self, board):
        """
        :type board: List[List[str]]
        :rtype: None Do not return anything, modify board in-place instead.
        """
        m = len(board[0])
        n = len(board)
        visited =[[0]*m for _ in range(n)]
        def solve(x,y):
            if x<0 or y<0 or x>=n or y>=m or board[x][y] != "O":
                return 
            board[x][y] = "S"
            solve(x+1,y)
            solve(x-1,y)
            solve(x,y+1)
            solve(x,y-1)

        for i in range(n):
            if board[i][0] == "O":
                solve(i,0)
            if board[i][m-1] == "O":
                solve(i,m-1)
        for j in range(m):
            if board[0][j]=="O":
                solve(0,j)
            if board[n-1][j]=="O":
                solve(n-1,j)
        for i in range(n):
            for j in range(m):
                if board[i][j] == "O":
                    board[i][j] = "X"
                if board[i][j] == "S":
                    board[i][j] = "O"
        return board
```
