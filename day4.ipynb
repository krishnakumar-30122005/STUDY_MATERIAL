# PERMUTATIONS

```python
nums = [1,2,3]
n = 3
res =[]
path =[]
visited = [0]*n
def solve():
    if len(res) == n:
        path.append(res[:])
        return
    for i in range(n):
        if visited[i]==1:
            continue 
        visited[i]=1
        res.append(nums[i])
        solve()
        visited[i]=0
        res.pop()
    return 
solve()
print(path)
```

# COMBINATIONS

```python
nums = [1,2,3]
k=2
res = []
path =[]
def solve(start):
    if len(res) == k:
        path.append(res[:])
        return 
    for i in range(start,n):
        res.append(nums[i])
        solve(i+1)
        res.pop()
    return 
solve(0)
print(path)
```

# MAZE PROBLEM

```python
maze = [
  [0,0,0],
  [1,1,0],
  [0,0,0]
]
start = (0,0)
end   = (2,2)
sx,sy = start
tx,ty = end
path =[]
res=[]
def solve(x,y):
  if x==tx and y == ty:
    path.append(res[:])
    return True 
  if x<0 or y<0 or x>=n or y>=n or maze[x][y] ==1 :
    return False
  res.append([x,y])
  if (solve(x+1,y)):
    return True
  if solve(x,y+1):
    return True
  res.pop()
  return False

print(solve(sx,sy))
print(path)
```

# KNIGHT PROBLEM

```python
n = 8
sx,sy = 0,0
tx,ty = 7,7
ans=[float('inf')]
visited = [[False]*n for _ in range(n)]
def solve(x,y,step):
    if x==tx and y==ty:
        if ans[0]>step:
            ans[0] = step
        return 
    if x<0 or y<0 or x>=n or y>=n or visited[x][y]==True:
        return
    if step>=ans[0]:
        return
    visited[x][y] = True
    solve(x+2,y+1,step+1)
    solve(x-2,y-1,step+1)
    solve(x+1,y+2,step+1)
    solve(x-1,y-2,step+1)
    solve(x+2,y-1,step+1)
    solve(x-2,y+1,step+1)
    solve(x+1,y-2,step+1)
    solve(x-1,y+2,step+1)
    visited[x][y] = False
    return 
solve(0,0,0)
print(ans[0])

```

# N QUEEN

```python
col =set()
s1 = set()
s2 = set()
n = 4
grid=[["."]*n for _ in range(n)]
res =[]
def solve(r):
    if r == len(grid):
        res.append([''.join(x) for x in grid])
    for c in range(n):
        if c in col or r+c in s1 or r-c in s2:
            continue
        grid[r][c] ="Q"
        col.add(c)
        s1.add(r+c)
        s2.add(r-c)
        solve(r+1)
        grid[r][c]="."
        col.remove(c)
        s1.remove(r+c)
        s2.remove(r-c)
    return
solve(0)
for row in res:
    print(*row)
    print("_________________________________________________")
```

# WORD SEARCH

```python
board = [
  ['A','B','C','E'],   
  ['S','F','C','S'], 
  ['A','D','E','E']
]
n = len(board)
m = len(board[0])
word = "ABCCED"
res =[]
res1=[]
path =[]
solution = [[0]*m for _ in range(n)]
def solve(x,y,res,idx):
    if res == word:
        path.append(res1[:]+[x,y])
        return True
    if x<0 or y<0 or x>=n or y>=m or solution[x][y] ==1 or board[x][y]!=word[idx]:
        return False
    solution[x][y] =1
    res1.append([x,y])
    if solve(x+1,y,res+word[idx],idx+1):
        return True
    if solve(x-1,y,res+word[idx],idx+1):
        return True
    if solve(x,y+1,res+word[idx],idx+1):
        return True
    if solve(x,y-1,res+word[idx],idx+1):
        return True
    res1.pop()
    solution[x][y] =0
    return False
for i in range(n):
    found = False
    for j in range(m):
        if solve(i,j,"",0):
            found = True
            print(found)
            break
    if found:
        break
print(path)
```
