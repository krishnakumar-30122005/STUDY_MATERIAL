# 1️⃣ 416 – Partition Equal Subset Sum

```python
class Solution(object):
    def canPartition(self, nums):
        """
        :type nums: List[int]
        :rtype: bool
        """
        def subset_sum(target):
            n = len(nums)
            dp = [[False]*(target+1) for _ in range(n+1)]

            for i in range(n+1):
                dp[i][0] = True

            for i in range(1,n+1):
                for j in range(1,target+1):
                    if nums[i-1]<=j:
                        dp[i][j] = dp[i-1][j] or dp[i-1][j-nums[i-1]]
                    else:
                        dp[i][j] = dp[i-1][j]
            return dp[n][target]
            
        sum1 = sum(nums)
        if sum1%2==1:
            return False
        else:
            return subset_sum(sum1//2)
```
