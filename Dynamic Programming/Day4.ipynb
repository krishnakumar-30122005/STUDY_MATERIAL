# Fibinoci series

```python
n = 5

dp = [0]*(n+1)

dp[0] = 0
dp[1] = 1

for i in range(2,n+1):
  dp[i]  = dp[i-1] + dp[i-2]
print(dp)
print(dp[n])
```

# Robber 

```python
class Solution(object):
    def rob(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """ 
        n = len(nums)
        dp= [0]*(n)
        if n ==1:
            return nums[0]
        dp[0] = nums[0]
        dp[1] = max(nums[0],nums[1])

        for i in range(2,n):
            dp[i] = max(dp[i-1],nums[i]+dp[i-2])

        return dp[-1]
```

# Grid Traversal

```python
def uniquePaths(m, n):
    dp = [[0]*n for _ in range(m)]

    # base case
    for i in range(m):
        dp[i][0] = 1
    for j in range(n):
        dp[0][j] = 1

    for i in range(1, m):
        for j in range(1, n):
            dp[i][j] = dp[i-1][j] + dp[i][j-1]

    return dp[m-1][n-1]
```

# CanSum

```python
value = [3,4,5,7]
target = 7

dp = [None]*(target+1)
dp[0] =0
for i in range(target+1):
    if dp[i]!= None:
        for num in value:
            mark = i+num
            if mark<=target:
                if dp[mark] == None:
                    dp[mark] = [[num]]
                else:
                    dp[mark].append([i,num])
print(dp)
```
