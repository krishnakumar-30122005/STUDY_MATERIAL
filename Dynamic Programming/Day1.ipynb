# Climbing stairs

```python
class Solution(object):
    def climbStairs(self, n):
        """
        :type n: int
        :rtype: int
        """
        if n==1 :
            return n
        dp = [0]*(n)
        dp[0] = 1
        dp[1] = 2
        for i in range(2,n):
            dp[i] = dp[i-1]+dp[i-2]
        return dp[n-1]
```

# 746. Min Cost Climbing Stairs

```python
class Solution(object):
    def minCostClimbingStairs(self, cost):
        """
        :type cost: List[int]
        :rtype: int
        """
        n = len(cost)
        dp = [0]*n
        dp[n-1] = cost[n-1]
        dp[n-2] = cost[n-2]

        for i in range(n-3,-1,-1):
            dp[i] = cost[i] + min(dp[i+1],dp[i+2])

        return min(dp[0],dp[1])

        
```

# 198. House Robber

```python
class Solution(object):
    def rob(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """ 
        n = len(nums)
        dp= [0]*(n)
        if n ==1:
            return nums[0]
        dp[0] = nums[0]
        dp[1] = max(nums[0],nums[1])

        for i in range(2,n):
            dp[i] = max(dp[i-1],nums[i]+dp[i-2])

        return dp[-1]
```

# 62. Unique Paths

```python
class Solution(object):
    def uniquePaths(self, m, n):
        """
        :type m: int
        :type n: int
        :rtype: int
        """
        dp =[[0]*(n+1) for i in range(m+1)]
        dp[1][1] = 1

        for i in range(1,m+1):
            for j in range(1,n+1):
                if i==1 and j==1:
                    continue
                else:
                    dp[i][j]  = dp[i-1][j]+dp[i][j-1]
        return dp[m][n]
```

# 64. Minimum Path Sum

```python
class Solution(object):
    def minPathSum(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """
        INF =  float('inf')
        n = len(grid)
        m = len(grid[0])
        dp = [[INF]*(m+1) for _ in range(n+1)]

        dp[1][1] = grid[0][0]

        for i in range(1,n+1):
            for j in range(1,m+1):
                if i==1 and j==1:
                    continue
                else:
                    dp[i][j] = min(dp[i-1][j],dp[i][j-1])+grid[i-1][j-1]
        return dp[n][m]
```
