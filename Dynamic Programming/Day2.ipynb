# SUBSEQUENCE SUM USING DYNAMIC PROGRAMMING

```python
def subsequence_sum(nums, target):
    n = len(nums)
    
    dp = [[False] * (target + 1) for _ in range(n + 1)]
    
    # base case
    for i in range(n + 1):
        dp[i][0] = True

    for i in range(1, n + 1):
        for s in range(1, target + 1):
            if nums[i - 1] <= s:
                dp[i][s] = dp[i - 1][s] or dp[i - 1][s - nums[i - 1]]
            else:
                dp[i][s] = dp[i - 1][s]

    return dp[n][target]
```
