# 0-1 knapsack problem 

```python
def knapsack(weights, values, W):
    n = len(weights)
    
    dp = [[0] * (W + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        for w in range(1, W + 1):
            if weights[i - 1] <= w:
                dp[i][w] = max(
                    dp[i - 1][w],
                    values[i - 1] + dp[i - 1][w - weights[i - 1]]
                )      
            else:
                dp[i][w] = dp[i - 1][w]

    return dp[n][W]
```

# Subset sum problem 

```python
arr = [2,3,7,8,10]
sum = 11
n = len(arr)
dp = [[False]*(sum+1) for i in range(n+1)]

for i in range(n+1):
    dp[i][0] = True

for i in range(n+1):
    for j in range(sum+1):
        if arr[i-1]<=j:
            dp[i][j] = dp[i-1][j] or dp[i-1][j - arr[i-1]]
        else:
            dp[i][j] = dp[i-1][j]
print(dp[n][sum])
```

# Count the no.of subset with the given difference

given a array and difference to find subset for using mathematical realtion 

s1-> subset1

s2-> subset2

we know that 
s1-s2 -> diff

s1+s2 -> sum(arr)

compute maths
we obtain

s1 = (diff+sum(arr))//2

we create a dp table for the index and s1
return the last index


```python
arr = [1,1,2,3]
diff = 1
n = len(arr)
sum1 = sum(arr)

sum2 = (diff+sum1)//2
dp = [[0]*(sum2+1) for _ in range(n+1)]

for i in range(n+1):
    dp[i][0] = 1

for i in range(1,n+1):
    for j in range(1,sum2+1):
        if arr[i-1]<=j:
            dp[i][j] = dp[i-1][j] + dp[i-1][j-arr[i-1]]
        else:
            dp[i][j] = dp[i-1][j]
print(dp[n][sum2])
```

# Target Sum 

This is also same as the no. of subset with the given difference since the sign are changed to get the targeted sum but there is a pattern lies inbetween them as you can see sum of s1 - sum of s2 always equal to the target

s1 = sum of all positive elements
s2 = sum of all negative elements 

s1-s2 = target
as same as difference


```python
arr = [1,1,1,1,1]
diff = 3
n = len(arr)
sum1 = sum(arr)

sum2 = (diff+sum1)//2
dp = [[0]*(sum2+1) for _ in range(n+1)]

for i in range(n+1):
    dp[i][0] = 1

for i in range(1,n+1):
    for j in range(1,sum2+1):
        if arr[i-1]<=j:
            dp[i][j] = dp[i-1][j] + dp[i-1][j-arr[i-1]]
        else:
            dp[i][j] = dp[i-1][j]
print(dp[n][sum2])
```

In the bounded knapsack items have limited quantiy but in the unbound knapsack the items does not have the limites quantity it can be used unlimitedly as long as needed

dp transition for un-bounded knapsack

dp[i][w] = max(

    dp[i-1][w],                 # exclude item

    value[i] + dp[i-1][w-weight[i]]  # include item ONCE
    
)
