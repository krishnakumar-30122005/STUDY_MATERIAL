# 0-1 knapsack problem 

```python
def knapsack(weights, values, W):
    n = len(weights)
    
    dp = [[0] * (W + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        for w in range(1, W + 1):
            if weights[i - 1] <= w:
                dp[i][w] = max(
                    dp[i - 1][w],
                    values[i - 1] + dp[i - 1][w - weights[i - 1]]
                )      
            else:
                dp[i][w] = dp[i - 1][w]

    return dp[n][W]
```

# Coin changes 1 

```python
class Solution(object):
    def coinChange(self, coins, amount):
        """
        :type coins: List[int]
        :type amount: int
        :rtype: int
        """
        INF = amount+1
        n = len(coins)
        dp = [[INF]*(amount+1) for _ in range(n+1)]

        for i in range(n+1):
            dp[i][0] = 0
        for i in range(1,n+1):
            for j in range(1,amount+1):
                if coins[i-1]<=j:
                    dp[i][j] = min(dp[i-1][j], 1+dp[i][j-coins[i-1]])
                else:
                    dp[i][j] = dp[i-1][j]
        return dp[n][amount] if dp[n][amount]!=INF else -1    
```

# LONGEST COMMON SUBSEQUENCE

longest sequence subsequence is the problem which contains two string we were asked to find the count of common element in the two string 

s1 = "abcdgfh"

s2 = "abdefhr"

the common subsequence are "abdh"

expected to return the count("abdh")

# Recursive solution


start from last index of both string , there are number of possible result 

if both are equal decrement both 

else

branch from x-1 and y-1 

```python
str1 = "abcdgh"
str2 = "abedfhr"

n = len(str1)
m = len(str2)

def solve(str1,str2,n,m):
    if n==0 or m==0:
        return 0
        
    if str1[n-1] == str2[m-1]:
        return 1+solve(str1,str2,n-1,m-1)
    else:
        return max(solve(str1,str2,n-1,m),solve(str1,str2,n,m-1))

ans = solve(str1,str2,n,m)
print(ans)
```

# Recursion with memoization

in rescursive method of solving the problem it will check whether the function is called before that if so fetch the answer from that part 


```python
str1 = "abcdgh"
str2 = "abedfhr"

n = len(str1)
m = len(str2)

dp = [[-1]*(m+1) for _ in range(n+1)]
def solve(str1,str2,n,m):
    if n == 0 or m == 0 :
        return 0
    if dp[n][m]!=-1:
        return dp[n][m]
    if str1[n-1] == str2[m-1]:
        dp[n][m] = 1+solve(str1,str2,n-1,m-1)
    else:
        dp[n][m] =  max(solve(str1,str2,n-1,m),solve(str1,str2,n,m-1))
    
    return dp[n][m]
ans =solve(str1,str2,n,m)
print(ans)
```

# Maximum Length of Repeated Subarray

```python
class Solution(object):
    def findLength(self, nums1, nums2):
        """
        :type nums1: List[int]
        :type nums2: List[int]
        :rtype: int
        """

        n = len(nums1)
        m = len(nums2)

        dp = [[0]*(m+1) for _ in range(n+1)]
        max_len = float('-inf')
        for i in range(1,n+1):
            for j in range(1,m+1):
                if nums1[i-1] == nums2[j-1]:
                    dp[i][j] = 1+dp[i-1][j-1]
                    max_len = max(max_len,dp[i][j])
                else:
                    dp[i][j] = 0
        return max_len if max_len!=float('-inf') else 0
```

# 583. Delete Operation for Two Strings

```python
class Solution(object):
    def minDistance(self, word1, word2):
        """
        :type word1: str
        :type word2: str
        :rtype: int
        """
        n=len(word1)
        m = len(word2)

        dp = [[0]*(m+1) for i in range(n+1)]

        for i in range(1,n+1):
            for j in range(1,m+1):
                if word1[i-1] == word2[j-1]:
                    dp[i][j] = 1+dp[i-1][j-1]
                else:
                    dp[i][j] = max(dp[i-1][j], dp[i][j-1])
        lcs = dp[n][m]
        return (n-lcs)+(m-lcs)
```
