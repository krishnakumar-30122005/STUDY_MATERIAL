# LEETCODE PROBLEM 51

```python
col = set()
sc = set()
sr = set()
n=4
grid = [["."]*(n) for _ in range(n)]
res = []
def solve(r):
    if r == n:
        res.append(["".join(x) for x in grid])
        return 
    for c in range(n):
        if c in col or r+c in sc or r-c in sr:
            continue
        grid[r][c] = "Q"
        col.add(c)
        sc.add(r+c)
        sr.add(r-c)
        solve(r+1)
        grid[r][c] = "."
        sc.remove(r+c)
        sr.remove(r-c)
        col.remove(c)
    return 
solve(0)
for row in res:
    print(*row)
```

# LEETCODE 2435
I tried to solve using backtracking but at some point is return TLE since the backtracking not work for large amount of data, Inother word the stack size is used 

```python
grid = [[5,2,4],[3,0,5],[0,7,2]]
k = 3
n = len(grid)
m = len(grid[0])
tx,ty = n-1,m-1
ans= [0]
solution = [[0]*m for _ in range(n)]

def solve(solution,x,y,s):
    if x == tx and y == ty:
        s+=grid[x][y]
        if s%k==0:
            ans[0]+=1
    if x<0 or y<0 or x>=n or y>=m or solution[x][y]==1:
        return 
    
    solution[x][y] = 1
    solve(solution,x+1,y,s+grid[x][y])
    solve(solution,x,y+1,s+grid[x][y])
    solution[x][y] = 0
    
    return

solve(solution,0,0,0)
print(ans[0])
```

# LEETCODE 1219
Path with maximum gold, I solved it using backtracking


```python
class Solution(object):
    def getMaximumGold(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """
        n = len(grid)
        m = len(grid[0])
        solution = [[0]*m for i in range(n)]
        ans = [0]
        def solve(solution,x,y,s):
            
            if s>ans[0]:
                ans[0]=max(ans[0],s)
            if x<0 or y<0 or x>=n or y>=m or grid[x][y] == 0 or solution[x][y]==1:
                return 
            solution[x][y]=1
            solve(solution,x+1,y,s+grid[x][y])
            solve(solution,x-1,y,s+grid[x][y])
            solve(solution,x,y+1,s+grid[x][y])
            solve(solution,x,y-1,s+grid[x][y])

            solution[x][y] = 0
            return 
        for i in range(n):
            for j in range(m):
                solve(solution,i,j,0)
        return ans[0]
            



        
```

# LEETCODE 200
Number of Islands problem solved using backtracking

```python
class Solution(object):
    def numIslands(self, grid):
        """
        :type grid: List[List[str]]
        :rtype: int
        """
        island = 0
        n= len(grid)
        m = len(grid[0])
        def solve(x,y):
            if x<0 or y<0 or x>=n or y>=m or grid[x][y]=="0":
                return
            grid[x][y] = "0"
            solve(x+1,y)
            solve(x-1,y)
            solve(x,y+1)
            solve(x,y-1)
            return

        for i in range(len(grid)):
            for j in range(len(grid[0])):
                if grid[i][j] == "1":
                    island+=1
                    solve(i,j)
        return island
```
