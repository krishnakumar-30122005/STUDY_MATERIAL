```python
arr = eval(input())
k = int(input())

res = []
for i in range(len(arr)-k+1):
    max_value = arr[i]
    for j in range(i,i+k):
        max_value = max(max_value,arr[j])
    res.append(max_value)
print(res)
```

```python
str1 = input().strip()
k = int(input())
count = {}
l = 0
ans = 0
max_freq = 0
for r in range(len(str1)):
    ch = str1[r]
    count[ch] = count.get(ch,0)+1
    max_freq = max(max_freq,count[ch])
    while (r-l-max_freq+1)>k:
        count[str1[l]]-=1
        l+=1
    ans = max(ans,r-l+1)
print(ans)
```

```python
from collections import Counter
s = input().strip()
t = input().strip()
need = Counter(t)
missing = len(t)

start = 0
l = 0
ans = float("inf")

for r,ch in enumerate(s):
    if need[ch]>0:
        missing-=1
    need[ch]-=1
    
    while missing == 0:
        if (r-l+1)<ans:
            ans = r-l+1
            start = l
        need[s[l]]+=1
        if need[s[l]]>0:
            missing+=1
        l+=1
print("" if ans == float('inf') else s[start:start+ans])
```

```python
str1 = input().strip()
seen = set()
l = 0
ans = 0
for r in range(len(str1)):
    ch = str1[r]
    while ch in seen:
        seen.remove(str1[l])
        l+=1
    seen.add(ch)
    ans = max(ans,r-l+1)
print(ans)
            
```
